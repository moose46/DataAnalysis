```python
import pandas as pd
df_sales = pd.read_excel("supermarket_sales.csv.xlsx")
# df_sales = pd.read_csv("supermarket_sales.csv"
df_sales.select_dtypes(exclude=["datetime64","datetime"])
df_sales.pivot_table(index="Gender",aggfunc="sum")
df_sales.pivot_table(index="Gender",values=["Quantity", "Total"],aggfunc="sum") # type: ignore
# df_sales.pivot_table(index="Gender",columns="Product line",values=["Total"],aggfunc="sum").to_csv("supermarket_sales_by_gender.csv") # type: ignore
df_sales.pivot_table(index="Gender",columns="Product line",values=["Total"],aggfunc="sum") # type: ignore
```

```python
import pandas as pd
df_population_raw = pd.read_csv("population_total.csv")
# dropping all null values
df_population_raw = df_population_raw.dropna()

# making a pivot table
df_pivot = df_population_raw.pivot(index="year",columns="country", values="population")
df_pivot = df_pivot[["United States","India","China"]]
df_pivot
```
